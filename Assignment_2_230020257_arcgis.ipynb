## 1.3 What's the purpose
In this exercise I described the purpose of the 1st 2 expressions in the code below



```python
import pandas as pd
import geopandas as gpd
dat = pd.read_csv('data/world_cities.csv')  ## Import CSV file
geom = gpd.points_from_xy(dat['lon'], dat['lat'])
geom = gpd.GeoSeries(geom)
dat = gpd.GeoDataFrame(dat, geometry = gpd.GeoSeries(geom), crs = 4326)
dat.to_file('output/world_cities.shp')      ## Export Shapefile
```

The first 2 expressions are used to import python libraries into the notebook. Pandas is used for working with tabular data and geopandas is used for working with geospatial data. As this code involvolves CSV files and shapefiles, it's useful to have both pandas and geopandas libraries so that we can work with different data types.

## 1.8 Create a script that generates a multiplication table
In this execise I created a script that created a multiplication table. I defined a range of 1 to 10 and then printed the results for different numbers

```python
def multab(n):  # defines function that creates a multiplication table for any number
    for i in range(1, 11):  # for loop for numbers between 1 and 10
        print(n, "x", i, "=", n * i)  


multab(4)  # gives multiplication table for number 4 multiplied by number between 1 and 10
multab(7)
multab(9)
```

## 1.9 Loops
In this exercise I created a list that gave the count for each lunch item using a for loop

```python
lunch = ["Salad", "Salad", "Egg", "Beef", "Potato", "Tea", "Chicken", "Potato", "Potato", "Coffee"]  # creates lunch list with items

item_count = {}  # empty dictionary for item counts
for item in lunch:  # for loop for lunch items
    if item not in item_count:
        item_count[item] = 1  # if item is not in the dictionary then give it value of 1
    else:
        item_count[item] += 1  # otherwise, keep adding 1 to the value for that item in the dictionary

print(item_count)  # prints item count results
```

## 2.12
I identified the maximum co2 emissions in the dataset and identified which food group it belonged to

```python
max = df["co2_emmission"].max()  # identifies max co2 values in the dataset
print(max)
max_row = df.loc[df["co2_emmission"] == max]  # select rows where co2 emissions are of the max value
print(max_row[["food_category", "country", "co2_emmission"]])  # prints the max row for co2 while selecting the rows "food_category, "country" and "co2 emmissions"
```

## 2.13
I identifed how many countries produced more than 1000 Kg CO2/person/year for at least one food type

```python
mask = df["co2_emmission"] > 1000  # creates a boolean mask for the co2 values greater than 1000
countries = set(df.loc[mask, "country"])
print("countries:", countries)  # prints countries of co2 values greater than 1000
print("number of countries:", len(countries))  # prints the number of countries that produce more than 1000kg of co2/person/year for at least one food type
```

## 2.15
I identifed the total emissions of all the meat products (Pork, Poultry, Fish, Lamb & Goat, Beef) in the dataset combined

```python
meat_items = ["Pork", "Poultry", "Fish", "Lamb & Goat", "Beef"]  # creates list of meat items
meat_items_df = df.query('food_category in @meat_items')  # filters the dataframe to include meat items from the list
total_emissions = meat_items_df["co2_emmission"].sum()  # calculates the total emissions from meat items 
print(total_emissions)  # prints the results
```

## 2.16
I identifed the total emissions of all other (non-meat) products in the dataset combined

```python
meat_items = ["Pork", "Poultry", "Fish", "Lamb & Goat", "Beef"]  # creates list of meat items
non_meat_items_df = df.query('food_category not in @meat_items')  # filters the dataframe to include only items not in the meat items list
total_emissions = non_meat_items_df["co2_emmission"].sum()  # calculates the total emission from non meat items
print(total_emissions)  # prints the results
```
